```python
filtered_df = filtered_df.copy()
# There was an error that kept popping up whenever I would try to modify the original filtered_df; it seems like it wants a copy, so I'll use a copy
filtered_df['updated'] = pd.to_datetime(filtered_df['updated'])
# Datetime conversion
exploded_df = filtered_df.explode('des_facet')
keyword_counts = exploded_df['des_facet'].value_counts().reset_index()
# Explode into rows and check counts
keyword_counts.columns = ['keyword', 'count']
# For the visualization component; I'd like to have the columns here.
top_keywords = keyword_counts.head(10)
# We want the top 10, so just the head(10) for me please.
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='count', y='keyword', data=top_keywords, palette='pastel')
# Creating a barplot using the top 10 keywords and checking their frequency
plt.title('Top 10 Most Frequently Mentioned Keywords')
plt.xlabel('Frequency')
plt.ylabel('Keyword')
plt.show()
```

```python
filtered_df['week'] = filtered_df['updated'].dt.to_period('W').apply(lambda r: r.start_time)
# Converts the updated date column to weeks. From what I've learned, the lambda function is sort of like the summation symbol
# I'm applying it to each start time entry of each week.
articles_per_week = filtered_df.groupby('week').size().reset_index(name='count')
# Basically just getting the actual count of articles per week
plt.figure(figsize=(10, 5))
sns.lineplot(x='week', y='count', data=articles_per_week, marker='o')
plt.title('Number of Articles Published Per Week')
plt.xlabel('Week')
plt.ylabel('Number of Articles')
plt.show()
```
